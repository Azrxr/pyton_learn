tipe data primitif
numbers : int, float, complex

tipe data primitif ke dua
boolean : true atau false

```python
int = 6
print(type(int))

float = 6.0
print(type(float))

complex = 10+2j
print(type(complex))

#type data numbers bersifat immutable (tidak dapat berubah)
var = 10
print(var)
print(id(var))

#bukan merubah nilai var tapi menambah objek baru
var = 100
print(var)
print(id(var))

#nilai dianggap false :
"""
Nilai yang sudah didefinisikan bernilai salah: None dan False.
Angka nol dari semua tipe numerik: 0, 0.0, 0j, Decimal(0), Fraction(0,1).
Urutan (sequence) dan koleksi (collection) yang kosong: “”, (), {}, set(), range(0).
"""

benar = True
print(benar)
print(type(benar))

salah = False
print(salah)
print(type(salah))
```

String

```python

singgle = 'singgle quote'
print(singgle)

double_quote = "string dengan double quote"
print(double_quote)

multi_line = """halo
Kapan terakhir kali kita bertemu?
Kita bertemu hari Jum’at yang lalu."""
print(multi_line)
print(type(multi_line))

#index: akses karakter setiap string
index = 'abcdefghij'
print(index[0])
#tidak bisa mengubah string di dalamnya (immutable)
#index[0] = 'b'

#indexing + slecing
print(index[2:])

#menampilkan text string berdasarkan input
name = "Kucing Garong"
#formated string
print(f"hello, meong merupakan: {name}")
#%-formating
print("meong merupakan %s" % (name))
#str.format
print("meong saya {}".format(name))
```

Tipe data collection

```python
#list tipe datanya bebas bisa di campur
x = [ 1, 0, "yuhu", False, 5.0]
print(x)
print(type(x))
#index
print(x[2])

#list bisa berubah (mutable)
x[0] = "Indonesia"
print(x)

#index slicing
#sequence[start:stop:step]
hw = ["laptop", "monitor", "mouse", "mousepad", "keyboard", "webcam", "microphone"]
print(hw[0:5:2])
print(hw[0:])
print(hw[:3])

#tuple : immutable
var_tuple = (1, "Oy", 1+2j)
print(var_tuple)
print(type(var_tuple))
#index dan slecing sama seperti list

#set {} : sekumpulan(bukan urutan) jadi tidak bisa di index
var_set = {1,2,3,4,5}
print(var_set)
print(type(var_set))

#set bersifat uniqe tanpa duplikat
var_set_dup = {1,1,2,2,4,4,5,6,7,8,8}
print(var_set_dup)

#set merupakan himpunan mtk
#union(gabungan) dan intersection(irisan)

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

union = set1.union(set2)
print("Union: ", union)

intersection = set1.intersection(set2)
print("Intersection: ", intersection)

#dictionary: key-value
var_dictionary = {'name': "wardah", 'age': 28, 'isMaried': False}
print(type(var_dictionary))
#untuk mengambil data harus tau 'key'
print(var_dictionary['name'])
#menambah
var_dictionary['job'] = "pemumulung"
print("tambah: ", var_dictionary)
#hapus
del var_dictionary['isMaried']
print("hapus: ", var_dictionary)
#update
var_dictionary['name'] = "liquid"
print("ubah: ", var_dictionary)

#konversi
#print(f"intToFloat: {float(5)}")
```
