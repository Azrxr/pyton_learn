One-liner

One-liner merupakan gaya penulisan pada Python yang memungkinkan Anda untuk membuat sebuah kode hanya dalam satu baris.

```python
x = 1
y = 2

print("sebelum pertukaran: ")
print("nilai x =", x)
print("nilai y =", y)

temp = x
x = y
y = temp

print("setelah pertukaran: ")
print("nilai x =", x)
print("nilai y =", y)

print("Teknik One Liner")
gelas1 = "air putih"
gelas2 = "jus mangga"

print("sebelum pertukaran: ")
print("isi gelas1 =", gelas1)
print("isi gelas2 =", gelas2)

print("setelah pertukaran: ")
gelas1, gelas2 = gelas2, gelas1
print("isi gelas1 =", gelas1)
print("isi gelas2 =", gelas2)
```
