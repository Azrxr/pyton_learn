Jenis-Jenis Ekspresi  
Pada dasarnya, jenis-jenis ekspresi dibagi menjadi dua. Pertama adalah menurut jumlah operan (arity) dari operator dan kedua adalah menurut tipe data yang dihasilkan.

Ekspresi Menurut Arity dari Operator  
Biner (memiliki 2 opran)= (x + y), (x - y), (x * y), (x / y), (x ** y), (x < y), (x <= y), (x > y), (x >= y), (x % y), (x == y), (x != y).  
uner (bentuk dasar operasi dengan 1 operan) = (x += 1), (x-=1), (not x).  



```python
#expresi uner
a = True
a = not a
print(a) #Variabel a bernilai True, jika kita memberikan ekspresi "not a" yang artinya "not True", hasil yang didapat adalah False. 

b = 6
b -= 1
print(b) #Variabel b bernilai 6, jika kita memberikan ekspresi "b -= 1" yang artinya "6 - 1", hasil yang didapat adalah 5.

c = 6
c += 1
print(c) #Variabel c bernilai 6, jika kita memberikan ekspresi "c += 1" yang artinya "6 + 1", hasil yang didapat adalah 7.

d = 10
print(-d) #Variabel d bernilai 10, jika kita memberikan ekspresi "-d" yang artinya "negasi dari 10", hasil yang didapat adalah -10.
```

Ekspresi Menurut Tipe Data yang Dihasilkan  
1. Ekspresi aritmetika: jenis ekspresi yang memiliki operan bertipe numerik dan menghasilkan numerik.  
Ekspresi aritmetika: numerik | operator | numerik = numerik  
2 + 2 = 4, 2 - 2 = 0  
2. Ekspresi relasional: jenis ekspresi yang memiliki operan bertipe numerik dan menghasilkan nilai boolean/logika.  
Ekspresi relasional: numerik | operator | numerik = boolean  
3 < 10 = True, 1 > 10 = False  
3. Ekspresi logika: jenis ekspresi yang memiliki operan bertipe boolean/logika dan menghasilkan nilai boolean/logika.  
Ekspresi logika: boolean | operator | boolean = boolean  
True or False = True


```python
print(2+2)
print(3<10)
print(True or False)
```
